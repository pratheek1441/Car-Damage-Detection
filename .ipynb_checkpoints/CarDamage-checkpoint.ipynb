```python
import yaml

# Define your data
data = {
    'train': r"C:\Users\prath\OneDrive\Documents\datasets\train\images",
    'val': r"C:\Users\prath\OneDrive\Documents\datasets\val\images",
    'test': r"C:\Users\prath\OneDrive\Documents\datasets\test\images",
    'nc': 7,  # Number of classes
    'names': ['crack_and_hole', 'medium_deformation', 'severe_deformation', 'severe_scratch', 'slight_deformation', 'slight_scratch', 'windshield_damage']  # Class names
}

# Write data to data.yaml file
with open('data.yaml', 'w') as file:
    yaml.dump(data, file)
```
